```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

cases = pd.read_csv("covid_19_india.csv")

# Checking the column names of cases file

cases.head(2)
```

```python
cases.info()
```

```python
cases.describe()
```

```python
cases.head(2)
```

```python
# creating a new column "Active" to see the active cases of each state
cases["Active"] = cases.Confirmed - (cases.Cured+cases.Deaths)
```

```python
cases.head(2)
```

```python
#renaming the state column for easy purpose
cases.rename(columns={"State/UnionTerritory":"State"},inplace=True)
```

```python
cases.head(2)
```

```python
# grouping the states in cases file to check the state wise death cases
# Here taking the max() because the count is the addition of each day

Deaths_Statewise= cases.groupby("State").Deaths.max().sort_values()
Deaths_Statewise
```

```python
#Plotting the deaths Vs state
Deaths_Statewise.plot(kind="bar",figsize=(15,5))
plt.xlabel("States")
plt.ylabel("No.of Deaths")
```

```python
#changing the date column type from object type to datetime
cases.Date = pd.to_datetime(cases.Date, dayfirst=True)
```

```python
#Analyzing the raise of cases on monthly basis
cases.set_index('Date',inplace=True)
cases['Confirmed'].resample(rule='M').sum().plot(figsize=(25,5))
```

```python
#Analyzing the raise of cases on quarterly basis
cases['Confirmed'].resample(rule='Q').sum().plot(figsize=(25,5))
plt.xlabel('Date')
plt.ylabel('count of Confirmed cases')
plt.show()
```

```python
cases['Confirmed'].plot(color='red')
plt.title('Confirmed cases to analyze the trend')
plt.show()
```

```python
#last date for which data is present
today = cases[cases.Date == '2021-08-11']
today
```

```python
#Sorting data with number of confirmed cases
max_confirmed_cases=today.sort_values(by="Confirmed",ascending=False)
max_confirmed_cases
```

```python
#Getting states with maximum number of confirmed cases
top_states_confirmed=max_confirmed_cases[0:10]

#Making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="State",y="Confirmed",data=top_states_confirmed,hue="State")
plt.show()
```

```python
#Sorting data with number of death cases
max_death_cases=today.sort_values(by="Deaths",ascending=False)
max_death_cases
```

```python
#Getting states with maximum number of death cases
top_states_death=max_death_cases[0:10]

#Making bar-plot for states with top death cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="State",y="Deaths",data=top_states_death,hue="State")
plt.show()
```

```python
#Sorting data w.r.t number of cured cases
max_cured_cases=today.sort_values(by="Cured",ascending=False)
max_cured_cases
```

```python
#Getting states with maximum number of cured cases
top_states_cured=max_cured_cases[0:10]

#Making bar-plot for states with top cured cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="State",y="Cured",data=top_states_cured,hue="State")
plt.show()
```

```python
d1=pd.read_csv('covid_19_india.csv')
#renaming the state column for easy purpose
d1.rename(columns={"State/UnionTerritory":"State"},inplace=True)

d1.State.dropna()
d2=d1[((d1.State=='Karnataka') | (d1.State=='Delhi') | (d1.State=='Madhya Pradesh')) & (d1.Date.str.contains('2020-05')|d1.Date.str.contains('2021-05'))]
d2['Month']=0
for i in d2.index:
    if '2020-05' in d2.Date[i]:
        d2['Month'][i]='May 2020'
    elif '2021-05' in d2.Date[i]:
        d2['Month'][i]='May 2021'
        
d2
sns.set(style="darkgrid")
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x='State',y='Deaths',data=d2,hue='Month',estimator=np.sum)
plt.show()
```
